```python
import pandas as pd
```

```python
pointers = {
    (-1, -1): "↖",  # NW
    (-1, 1): "↗",  # NE
    (1, 1): "↘",  # SE
    (1, -1): "↙",  # SW
    (0, 1): "→",  # Right
    (1, 0): "↓",  # Down
    (0, -1): "←",  # Left
    (-1, 0): "↑",  # Up
}
directions = tuple(pointers.keys())
```

```python
# Example matrix and word
matrix = [
    "HGIELSESROHAYCC",
    "HSCARFZHATSOLRD",
    "TAGLOVESINNGLTO",
    "EMISTLETOEDIICE",
    "ETFFERRWEYNTHST",
    "TSORFUFDDRAICNE",
    "GICICLESEASONOS",
    "NRTHTEAHTUTERIC",
    "IHTLIOCISNRZITC",
    "RCAEOISVOAOEMAE",
    "ESLLTINERJNEGTD",
    "TBGCOLDRFOORTIE",
    "TICINESSESMFHPC",
    "AFIREINSMTIIUIE",
    "HSELAGMMOECTECM",
    "CKRARCTICWINDEB",
    "AIETAKSNEDWAIRE",
    "NTERSRAEBRALOPR",
]
```

```python
df_map = pd.DataFrame([list(m) for m in matrix])
df_map
```

```python
word_list = [
    "ANTIFREEZE",
    "ARCTIC WIND",
    "ASTRONOMIC",
    "BECOME FROSTED",
    "CHATTERING TEETH",
    "CHILLY",
    "CHRISTMAS",
    "COLD",
    "DECEMBER",
    "FIRE",
    "FREEZE",
    "FROST",
    "GALE",
    "GLOVES",
    "HAT",
    "HORSE SLEIGH",
    "ICE",
    "ICICLE",
    "ICINESS",
    "IGLOO",
    "JANUARY",
    "MISTLETOE",
    "POLAR BEARS",
    "PRECIPITATION",
    "SALT",
    "SCARF",
    "SEASON",
    "SHIVER",
    "SKATE",
    "SKI",
    "SNOW",
    "STEW",
]
print(len(word_list))
```

```python
def search_word_in_matrix(matrix: list, word: str) -> list | None:
    """Find word in matrix."""
    rows = len(matrix)
    cols = len(matrix[0])

    def is_valid(x: int, y: int) -> bool:
        return 0 <= x < rows and 0 <= y < cols

    # Function to check if a word exists starting from (i, j) in any direction
    def check_word(i: int, j: int) -> int:
        for nth, direction in enumerate(directions):
            x, y = i, j
            try:
                for k in range(1, len(word)):
                    x += direction[0]
                    y += direction[1]
                    if not is_valid(x, y) or matrix[x][y] != word[k]:
                        raise IndexError  # noqa: TRY301
                return nth  # Found the word  # noqa: TRY300
            except IndexError:
                pass  # Index out of bounds, move to next direction

        return -1

    # Main loop: check each cell in the matrix as the start of the word
    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] == word[0] and (nth := check_word(i, j)) != -1:
                return (i, j, nth, len(word))  # Return the position of the start of the word
    return None  # Word not found


results = []
cnt = 0
for word in word_list:
    result = search_word_in_matrix(matrix, word.replace(" ", ""))
    if result:
        results.append(result)
        print(f"Word {word} found at: {result}" if result else "Word not found", pointers[directions[result[2]]])
    else:
        print(word)
    cnt += 1 if result else 0
print(cnt, len(word_list))
```

```python
for x, y, nth, length in results:
    direction = directions[nth]
    for _ in range(length):
        df_map.iloc[x, y] = "*"
        x += direction[0]  # noqa: PLW2901
        y += direction[1]  # noqa: PLW2901

df_map
```

```python
ans = []
for i, word in enumerate(word_list):
    result = results[i]
    direction = directions[result[2]]
    print(pointers.get(direction), word, result[0], result[1])
    ans.append((word, pointers.get(direction)))
```

```python
df_ans = pd.DataFrame(ans, columns=["word", "direction"])
df_ans.set_index("word", inplace=True)
df_ans["word_length"] = df_ans.index.str.len()
df_ans
```

```python
df_ans.groupby("direction").count().sort_values("word_length")
```

```python
df_ans = df_ans.reset_index()
df_ans.groupby("direction").agg(
    {
        "word": lambda x: ", ".join(x),  # Concatenate words in each group
        "word_length": "count",  # Count occurrences
    }
)
```
